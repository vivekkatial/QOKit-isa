```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)

n_samples = 1000
distributions = {
    'Uniform': np.random.uniform(-1, 1, n_samples),
    'Uniform+': np.random.uniform(0, 1, n_samples),
    'Normal': np.random.normal(0, 1, n_samples),
    'Exponential': np.random.exponential(1/0.2, n_samples),
    'Cauchy': np.random.standard_cauchy(n_samples),
    'Log-normal': np.random.lognormal(0, 1, n_samples),
    'No weights': np.ones(n_samples)
}


data = pd.DataFrame({k: v for k, v in distributions.items()})
melted_data = data.melt(var_name='Distribution', value_name='Value')

sns.set(style="whitegrid")
g = sns.FacetGrid(melted_data, col="Distribution", col_wrap=3, sharex=False, sharey=False)
g.map(sns.histplot, 'Value', kde=True, bins=30, color="skyblue")

g.set_titles("{col_name}")
g.set_axis_labels("Value", "Frequency")
plt.tight_layout()
plt.savefig("edge_weight_distributions.png", dpi=300)
plt.show()
```
